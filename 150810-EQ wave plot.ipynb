```python
#Using STIX Font (which similar to Times New Roman)
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

fig = plt.figure(figsize=(16,4), dpi=200)
axes = fig.add_axes([0.1, 0.15, 0.80, 0.75])# Setting left, bottom, width, height (range 0 to 1)

axes.plot(t,f9xm,'#80FF00',label="Measured $F9^2$")# Notice that Latex math sign are fully supported!
#axes.plot(t,f9xm,'g',label="Measured F9")
axes.plot(t,f9xa,'#00FFFF',label="Analyzed F9")# Color could be assign by html color code.
axes.plot(t,b1xm,'#FF00FF',label="Measured B1")# Place weaker signal at the front

#Set Legend
axes.legend(loc=1,ncol=1)

#Set Title and Axis Labels 
axes.set_xlabel('Time (s)')
axes.set_ylabel('Velocity (m/s)')
axes.set_title('EQ Simulation Results');

#Set Y-Axis Ticker Scientific Annotation 
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1)) 
axes.yaxis.set_major_formatter(formatter)

#Set Grid Line
axes.grid(color='#A4A4A4',# One may define the color by HTML code code
          alpha=1,
          linestyle='dashed',
          linewidth=0.5)

#Save the result in to the same folder
macbookPPI=227 
fig.savefig("filename.png",dpi=macbookPPI)# Saving the picture as PNG file
fig.savefig("filename.jpg",dpi=macbookPPI)#
fig.savefig("filename.pdf",dpi=macbookPPI)# Sometime PDF file is much favorable

```
